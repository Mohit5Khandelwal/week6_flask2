Q1. Explain GET and POST methods.

GET and POST are two commonly used HTTP methods for sending data from a client (such as a web browser) to a server. Here's an explanation of each:

1. GET method:
- GET is an HTTP method used to retrieve data from a server.
- When a client sends a GET request, it appends the data to be sent as parameters in the URL's query string.
- GET requests are typically used for retrieving data or performing read-only operations.
- GET requests are idempotent, meaning multiple identical GET requests should produce the same result and have no side effects on the server.
- Examples of GET requests include fetching a webpage, retrieving a list of articles, or retrieving specific data based on query parameters.

2. POST method:
- POST is an HTTP method used to submit data to a server.
- When a client sends a POST request, the data is included in the request body instead of the URL's query string.
- POST requests are commonly used for submitting form data, creating new resources, or performing actions that modify server data.
- POST requests are not idempotent, meaning multiple identical POST requests can have different effects on the server each time they are executed.
- Examples of POST requests include submitting a registration form, creating a new user account, or adding a comment to a blog post.

Q2. Why is request used in Flask?

In Flask, the request object is used to handle incoming HTTP requests from clients. It provides access to the data sent by the client, such as form data, query parameters, headers, and more. The request object allows Flask applications to extract and process the data sent by clients and generate appropriate responses.


Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to redirect the client to a different URL. It is commonly used to redirect users to a different page or route within the application after processing a request.


Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to separate the presentation logic from the application logic. Templates provide a way to generate dynamic HTML or other types of responses by combining static content with dynamic data. They allow you to define the structure and layout of your web pages and populate them with data from your Flask application.
